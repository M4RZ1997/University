```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = plt.imread('cameraman.tif')
img_sp = plt.imread('cameraman_sp.png')
```

```python
img.shape, img_sp.shape
```

```python
plt.imshow(img_sp, cmap='gray')
```

```python
def mean_filter(img, size=5):
    filtered = np.zeros_like(img)
    H, W = img.shape
    # implement mean filter
    for i in range(size//2, H-size//2): #valid
        for j in range(size//2, W-size//2):
            patch = img[i - size//2: i + size//2 + 1, j - size//2: j + size//2 + 1]
            filtered[i,j] = np.mean(patch)
    return filtered
```

```python
def median_filter(img, size=5):
    filtered = np.zeros_like(img)
    H, W = img.shape
    # implement median filter
    for i in range(size//2, H-size//2): #valid
        for j in range(size//2, W-size//2):
            patch = img[i - size//2: i + size//2 + 1, j - size//2: j + size//2 + 1]
            filtered[i,j] = np.median(patch)
    return filtered
```

```python
mean_filtered = mean_filter(img_sp, 5)
```

```python
plt.imshow(mean_filtered, cmap='gray')
plt.title('mean filter');
```

```python
median_filtered = median_filter(img_sp, 5)
```

```python
plt.imshow(median_filtered, cmap='gray')
plt.title('median filter');
```

```python
plt.imshow(img, cmap='gray')
plt.title('original image');
```

```python
def median_filter_improved(img, size=5):
    filtered = np.zeros_like(img)
    H, W = img.shape
    # implement improved median filter
    for i in range(size//2, H-size//2): #valid
        for j in range(size//2, W-size//2):
            patch = img[i - size//2: i + size//2 + 1, j - size//2: j + size//2 + 1]
            if img[i,j] <= np.percentile(patch, 10) or img[i,j] >= np.percentile(patch, 90):
                filtered[i,j] = np.median(patch)
            else:
                filtered[i,j] = img[i,j]
    return filtered
```

```python
improved_median_filtered = median_filter_improved(img_sp, 5)
```

```python
plt.imshow(improved_median_filtered, cmap='gray')
plt.title('improved median filter');
```

```python
plt.imshow(median_filtered, cmap='gray')
plt.title('median filter');
```
