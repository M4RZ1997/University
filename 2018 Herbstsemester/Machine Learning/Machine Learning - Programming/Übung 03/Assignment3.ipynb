* What are advantages and disadvantages of MoG vs. K-means?

***Your Answer:***
Compared to K-Means MOG is much more flexible and uses more features to calculate the clusters. For example it also uses the unconstrained covariance structure of the clusters to assign the points to their center point. On the other hand K-Means uses only the euclidian distance for calculation. So for non-spherical data MOG performs much better than K-Means.

As K-Means only uses the euclidian distance to calculate the clusters it is much more likely that the same data set will lead to completely different outcomes as seen in our task. Also it does not perform very well if the clusters are of different size or point density.

In general K-Means computes the clusters and center points much faster than MOG due to the less implementation of different features.
